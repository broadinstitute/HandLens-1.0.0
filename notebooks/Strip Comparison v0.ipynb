```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
# First, we read the source (test) and control (destination) images, 
# and provide the corners of the strips manually.

# Read source image.
im_src = cv2.imread('strip-positive.png', cv2.IMREAD_GRAYSCALE)
# Four corners of the book in source image
pts_src = np.array([[95, 149], [205, 150], [201, 2237],[90, 2237]], dtype=float)

# Read destination image.
im_dst = cv2.imread('strip-control.png', cv2.IMREAD_GRAYSCALE)
# Four corners of the book in destination image.
pts_dst = np.array([[89, 120],[195, 120],[172, 2202],[65, 2202]], dtype=float)
```

```python
# Calculate Homography between source (test) to destination (control) they are aligned

h, status = cv2.findHomography(pts_src, pts_dst)
    
# Warp source image to destination based on homography
im_out = cv2.warpPerspective(im_src, h, (im_dst.shape[1],im_dst.shape[0]))    
```

```python
plt.imshow(im_src, cmap='gray')
plt.title("Source Image")
plt.show()
```

```python
plt.imshow(im_dst, cmap='gray')
plt.title("Destination Image")
plt.show()
```

```python
plt.imshow(im_out, cmap='gray')
plt.title("Warped Source Image")
plt.show()
```

```python
outFilename = "aligned.jpg"
print("Saving aligned image : ", outFilename);     
cv2.imwrite(outFilename, im_out)
```

```python
# Now, let's compute a second homography to keep only the strip pixels, already aligned.

pts_ref = np.array([[0, 0],[100, 0],[100, 2100],[0, 2100]], dtype=float)

h, status = cv2.findHomography(pts_dst, pts_ref)

im_control_norm = cv2.warpPerspective(im_dst, h, (100, 2100))
im_test_norm = cv2.warpPerspective(im_out, h, (100, 2100))
```

```python
plt.imshow(im_control_norm, cmap='gray')
plt.title("Normalized Control Image")
plt.show()
```

```python
plt.imshow(im_test_norm, cmap='gray')
plt.title("Normalized Test Image")
plt.show()
```

```python
# Crop out the top and the bottom parts of the strip, and applying bilateral filtering for smoothing
# https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html

crop = im_control_norm[735:1200, 0:100]
im_control_norm_cropped = cv2.bilateralFilter(crop, 9, 75, 75)
# im_control_norm_cropped = cv2.cvtColor(crop, cv2.COLOR_RGB2GRAY)
    
crop = im_test_norm[735:1200, 0:100]
im_test_norm_cropped = cv2.bilateralFilter(crop, 9, 75, 75)
# im_test_norm_cropped = cv2.cvtColor(crop, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(im_control_norm_cropped, cmap='gray')
plt.title("Normalized+Cropped Control Image")
plt.show()
```

```python
plt.imshow(im_test_norm_cropped, cmap='gray')
plt.title("Normalized+Cropped Test Image")
plt.show()
```

```python
# Convert from uint8 to int32 so we can get negative values
im_test_norm_cropped = im_test_norm_cropped.astype('int32') 
im_control_norm_cropped = im_control_norm_cropped.astype('int32') 
```

```python
# Finally, we calculate the difference between the two aligned, cropped, smoothed strip images
# take the mean along the vertical direction, and plot the result. A positive test should present a
# negative peak at the beginning (indicating a positive signal) and a positive peak at the end 
# (indicating lack of control signal)

diff = im_test_norm_cropped - im_control_norm_cropped
nrows = diff.shape[0]
ncols = diff.shape[1]
values = np.array([])

# Another possibility could be to use OpenCV's image integrals:
# http://www.aishack.in/tutorials/integral-images-opencv/

for r in range(0, nrows):
    values = np.append(values, np.mean(diff[r]))
    
plt.plot(values)
```
